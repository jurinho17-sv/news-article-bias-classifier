```python
import pandas as pd

df = pd.read_csv("preprocess1.csv", engine='python')
df.sample(25)
```

```python
df = df.drop(columns=["Unnamed: 0","year","year_month","Date"])
```

```python
import re
import pandas as pd # Added for pd.isna()

def clean_text(text):
  text = re.sub(r"http\S+", '', text)
  text = re.sub(r"\n", ' ', text)
  text = re.sub(r"--+", '', text)
  text = re.sub(r"\s+", ' ', text)
  return text

def clean_url(url):
  # Handle potential NaN values
  if pd.isna(url):
    return ''

  # Ensure the url is a string before applying regex operations
  url = str(url)

  url = re.sub(r"https://", '', url)
  url = re.sub(r"http://", '', url)
  url = re.sub(r"www.", '', url)
  url = re.sub(r"\.\S*", '', url)
  url = re.sub(r"\s.*", '', url)
  return url

df["text"] = df["text"].apply(clean_text)

# Corrected column assignment syntax
df["source"] = df["url"].apply(clean_url)
df = df.drop(columns=["url"])
df.sample(25)
```

```python
df["first_three"] = df['text'].str.split().str[:3].str.join(' ')
```

```python
df = df[df['first_three'] != "no story found"]
```
