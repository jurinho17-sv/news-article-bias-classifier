```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("articoder/news-dataset-for-news-bias-analysis")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

file_path = os.path.join(path, "news_article_dataset.tsv")
df = pd.read_csv(file_path, sep="\t", encoding="utf-8")
print(df.shape)
print(df.columns)
df.head(10)
```

```python
# consolidating each article to its own row
text_bias = ["left_story_text", "center_story_text", "right_story_text"]
dftext = df.melt(id_vars = ["Topics","Date"], value_vars = text_bias, var_name = "text_label", value_name = "text")
dftext.sample(25)
```

```python
# same thing with urls
url_bias = ["left_story_url", "center_story_url", "right_story_url"]
dfurl = df.melt(id_vars = ["Topics","Date"], value_vars = url_bias, var_name = "url_label", value_name = "url")
dfurl.sample(25)
```

```python
# merging text and urls
full_df = pd.merge(dftext,dfurl,on=["Topics", "Date"])
full_df.tail(25)
```

```python
# labels are leans now for easier reading
full_df["lean"] = full_df["text_label"].str.split("_").str[0]
full_df["lean2"] = full_df["url_label"].str.split("_").str[0]

print(len(full_df))
full_df.head()
```

```python
# I believe each text is now assigned to all three urls from og dataframe?
# 89910 / 29970 = 3 (exactly)
# so we are making sure urls and text match
full_df = full_df[full_df["lean"] == full_df["lean2"]]
print(len(full_df))
```

```python
# lean is now consolidated; label encoding later?
full_df.drop(["url_label","text_label","lean2"], axis=1, inplace=True)
full_df.head(25)
```

```python
full_df["lean"].unique() # sanity check
```

```python
# checking nas in text
full_df["text_len"] = full_df["text"].str.len()

print("With na: " + str(len(full_df["text_len"])))
print("Without na: " + str(len(full_df["text_len"].dropna())))

# need to remove nas from entire dataframe
full_df.dropna(subset=['text_len'], inplace=True)
print("New dataframe: " + str(len(full_df["text_len"])))

full_df["text_len"].value_counts().sort_index(ascending=False)
```

```python
# Visualization 1: Histogram - Text Length
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data = full_df, x = "text_len")
plt.axvline(x = np.median(full_df["text_len"]), color = "red")
plt.title("Text Length of Dataset Articles")
plt.xlabel("Text Length")
plt.ylabel("Articles")
plt.show()
```

```python
# ooh, duplicates. remove
full_df.drop_duplicates(subset = "text", inplace = True)
print(full_df.shape)
full_df.head(25)
```

```python
# arbitary baseline, but these articles are probably not long enough for nlp

full_df[full_df["text_len"] < 25]
```

```python
full_df = full_df[full_df["text_len"] > 25]
```

```python
type(full_df.iloc[0]["Date"]) # is the date in datetime format?
```

```python
# no, so let's convert it
full_df["Date"] = pd.to_datetime(full_df["Date"])

full_df.head()
```

```python
type(full_df.iloc[0]["Date"]) # is the date in datetime format?
```

```python
# ok, now we can do some time-wise visualization
# but first let's get the year so we can do a line plot
full_df["year"] = full_df["Date"].dt.year
full_df.head()
```

```python
full_df["year"].value_counts().sort_index()
```

```python
# curious to see if we can get even more interesting results with year-month
full_df["year_month"] = full_df["Date"].dt.strftime('%Y-%m')
full_df["year_month"].value_counts().sort_index()
```

```python
# yes this is interesting, no need for year anymore
full_df.drop(["year"], axis=1)
```

```python
# to plot we actually need to see the value_counts data
ym_data = full_df["year_month"].value_counts().sort_index().to_frame()
ym_data
```

```python
# Visualization 2: Line Plot - Release Date
import matplotlib.dates as mdates

# explicitly convert the index to datetime even though I thought I already did that
ym_data.index = pd.to_datetime(ym_data.index)

# plot
sns.relplot(data=ym_data, x="year_month", y="count", kind="line", aspect=2)

# explicitly get the axes object
ax = plt.gca()

# year ticks
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# month ticks
ax.xaxis.set_minor_locator(mdates.MonthLocator())

plt.title("Release Date of Dataset Articles")
plt.xlabel("Year")
plt.ylabel("Articles")
plt.grid()
plt.show()
```

```python
# The last visualization I want to see is the proportion of political biases represented in the dataset
# which is probably best done in a pie chart

# seeing the topics would probably also be cool

lean_data = full_df["lean"].value_counts().to_frame()
lean_data
```

```python
topic_data = full_df["Topics"].value_counts().to_frame()
topic_data # wow how surprising "Politics" is first
```

```python
# There's 126 of these so let's see if we can cut some of the least occurring ones
topic_data = topic_data[topic_data["count"] > 700]
print(len(topic_data))
topic_data
```

```python
# Visualization 3: Pie Charts - Political Bias and Largest Topics
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,8))
plt.subplots_adjust(wspace=0.5)

ax1.pie(lean_data["count"], explode=(0.01,0.01,0.01), labels=lean_data.index, colors=['blue', 'red', 'gray'], autopct='%1.1f%%')
ax1.set_title("Political Biases in Dataset")

def absolute_value(val):
    a = round(val/100.*sum(topic_data["count"])) # Calculate the actual value from the percentage
    return f'{a:.0f}' # Format as a whole number

ax2.pie(topic_data["count"], labels=topic_data.index, explode=[0.01]*len(topic_data), autopct=absolute_value)
ax2.set_title("Largest Topics in Dataset")

plt.show()
```
